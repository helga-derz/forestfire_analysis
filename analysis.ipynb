# Описание данных

1. X - x-axis spatial coordinate within the Montesinho park map: 1 to 9
2. Y - y-axis spatial coordinate within the Montesinho park map: 2 to 9
3. month - month of the year: "jan" to "dec" 
4. day - day of the week: "mon" to "sun"
5. FFMC - FFMC index from the FWI system: 18.7 to 96.20
6. DMC - DMC index from the FWI system: 1.1 to 291.3 
7. DC - DC index from the FWI system: 7.9 to 860.6 
8. ISI - ISI index from the FWI system: 0.0 to 56.10
9. temp - temperature in Celsius degrees: 2.2 to 33.30
10. RH - relative humidity in %: 15.0 to 100
11. wind - wind speed in km/h: 0.40 to 9.40 
12. rain - outside rain in mm/m2 : 0.0 to 6.4 
13. area - the burned area of the forest (in ha): 0.00 to 1090.84 
(this output variable is very skewed towards 0.0, thus it may make
sense to model with the logarithm transform).

# Приступаем к методу QDA - квадратичному дискриминантному анализу

```python
t = [0.1*10**(-6)]
for i in range(10):
    t.append(t[i]*10)
qda_model = discriminant_analysis.QuadraticDiscriminantAnalysis()
gridQDA = GridSearchCV(qda_model, param_grid = {'reg_param':np.arange(0,1,0.01),'tol' : t}, cv=10)
gridQDA.fit(X, y)

best_cv_err = 1 - gridQDA.best_score_
best_reg_param = gridQDA.best_estimator_.reg_param
best_tol = gridQDA.best_estimator_.tol
print(best_cv_err, best_reg_param, best_tol)
```

```python
gridQDA.grid_scores_
lines = [[] for i in range(11)]
labels = [str(x) for x in t]
for x in gridQDA.grid_scores_:
    lines[t.index(x[0]['tol'])].append((x[0]['reg_param'],x[1]))
#print(lines)

for i in range(len(lines)):
    plt.plot([x[0] for x in lines[i]],[1 - x[1] for x in lines[i]], label = labels[i])
plt.legend()    
plt.title('CV Error on QDA\n')
plt.xlabel('reg_param')
plt.ylabel('error')
```
