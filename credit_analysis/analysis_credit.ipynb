# определение бинарных\небинарных признаков среди категориальных

```python
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
print(binary_columns, nonbinary_columns)
```

# замена бинарных категориальных признаков на 1 и 0

```python
data_binary = data[binary_columns]
#print(data_binary.describe())
for c in binary_columns:
    #print(c)
    for i in data_binary.index:
        if data_binary.at[i,c] == data_describe[c]['top']:
            data_binary.at[i,c] = 1
        else:
            data_binary.at[i,c] = 0

    data_binary[c] = data_binary[c].astype(int)
```

```python
data_binary.describe()
```

# Приступаем к методу QDA - квадратичному дискриминантному анализу

```python
t = [0.1*10**(-6)]
for i in range(10):
    t.append(t[i]*10)
qda_model = discriminant_analysis.QuadraticDiscriminantAnalysis()
gridQDA = GridSearchCV(qda_model, param_grid = {'reg_param':np.arange(0,1,0.01),'tol' : t}, cv=10)
gridQDA.fit(X, y)

best_cv_err = 1 - gridQDA.best_score_
best_reg_param = gridQDA.best_estimator_.reg_param
best_tol = gridQDA.best_estimator_.tol
print(best_cv_err, best_reg_param, best_tol)
```

```python
gridQDA.grid_scores_
lines = [[] for i in range(11)]
labels = [str(x) for x in t]
for x in gridQDA.grid_scores_:
    lines[t.index(x[0]['tol'])].append((x[0]['reg_param'],x[1]))
#print(lines)

for i in range(len(lines)):
    plt.plot([x[0] for x in lines[i]],[1 - x[1] for x in lines[i]], label = labels[i])
plt.legend()    
plt.title('CV Error on QDA\n')
plt.xlabel('reg_param')
plt.ylabel('error')
```

### Регулируя оценку ковариаций параметром reg_param, мы снижаем ошибку. Можно заметить, что ошибка не зависит от параметра tol, поэтому дальше мы будем использовать его значение по умолчанию.

# Метод Random Forest

### Decision Trees являются хорошим семейством базовых классификаторов для бэггинга, поскольку они достаточно сложны и могут достигать нулевой ошибки на любой выборке. Метод случайных подпространств позволяет снизить коррелированность между деревьями, избежав таким образом переобучения. Таким образом, случайный лес — это бэггинг над решающими деревьями, при обучении которых для каждого разбиения признаки выбираются из некоторого случайного подмножества признаков.

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=1000)
rf_model.fit(X_train, y_train)
```

```python
err_train = np.mean(y_train != rf_model.predict(X_train))
err_test = np.mean(y_test != rf_model.predict(X_test))
print(err_train, err_test)
```

```python
rf_model = RandomForestClassifier()
gridrf = GridSearchCV(estimator=rf_model, param_grid={'n_estimators':range(1,1000, 50)}, cv= 5)
gridrf.fit(X, y)
best_cv_err = 1 - gridrf.best_score_
best_n_estimators = gridrf.best_estimator_.n_estimators
print(best_cv_err, best_n_estimators)
```

```python
plt.plot(range(1,1000, 50), [1-x[1] for x in gridrf.grid_scores_])
plt.title('Error on Random Forest\n')
plt.xlabel('n_estimators')
plt.ylabel('error')
```

```python
n_estimators = range(1,1000, 50)

train_error = []
test_error = []
oob_error = []
for n in n_estimators:
    rf_model = RandomForestClassifier(n_estimators=n, oob_score = True, bootstrap = True)
    rf_model.fit(X, y)
    oob_error.append(1 - rf_model.oob_score_)
    train_error.append(np.mean(y_train != rf_model.predict(X_train)))
    test_error.append(np.mean(y_test  != rf_model.predict(X_test)))
    
    
plt.plot(n_estimators, train_error, label = 'train error')
plt.plot(n_estimators, test_error, label = 'test error')
plt.plot(n_estimators, oob_error, label = 'OOB error')
plt.legend()
plt.title('Train, Test and OOB errors on Random Forest\n')
plt.xlabel('n_estimators')
plt.ylabel('error')
```

# Метод Extreme Random Forest

### В сверхслучайных деревьях (Extremely Randomized Trees) больше случайности в том, как вычисляются разделения в узлах. Это обычно позволяет немного уменьшить дисперсию модели за счет несколько большего увеличения смещения.

# Gradient Boosted Trees

### GBM - это метод, который основывается на ансамбле классификаторов. Идея состоит в том, чтобы следующий добавленный классификатор был подготовлен для улучшения уже подготовленного ансамбля. В сранение: что для каждой каждый классификатор RF и ERT обучается независимо от остальных.

```python
from sklearn.ensemble import GradientBoostingClassifier
gbt_model = GradientBoostingClassifier()
gbt_model.fit(X_train, y_train)
```

```python
err_train = np.mean(y_train != gbt_model.predict(X_train))
err_test = np.mean(y_test != gbt_model.predict(X_test))
print(err_train, err_test)
```

```python
gridgbt = GridSearchCV(estimator=gbt_model, param_grid={'n_estimators':range(1,1000, 50)}, cv= 5)
gridgbt.fit(X, y)
best_cv_err = 1 - gridgbt.best_score_
best_n_estimators = gridgbt.best_estimator_.n_estimators
print(best_cv_err, best_n_estimators)
```

```python
plt.plot(range(1,1000, 50), [1-x[1] for x in gridgbt.grid_scores_])
plt.title('Error on GBD\n')
plt.xlabel('n_estimators')
plt.ylabel('error')
```
